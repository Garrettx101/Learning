## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as mp
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

## Import Data

```python
SPY = pdr.get_data_yahoo("SPY", start = "1975-01-01", end="2018-01-01")
```

```python
SPY = pd.DataFrame(SPY)
```

```python
print(SPY)
```

## Graphing

```python
SPY.plot(fontsize = 12)
```

```python
SPY_Close = SPY['Close']
```

```python
SPY_Close.plot()
```

## Function

```python
def plot_data(symbol, title = "Stock Prices"):
    ax = symbol.plot(fontsize = 12)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
```

```python
plot_data(SPY)
```

```python
plot_data(SPY["Close"])
```
