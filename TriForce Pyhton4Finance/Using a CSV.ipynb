## Import Mods

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import Data

```python
Data = pd.read_csv("E:\yFinance Data\^GSPC.csv", index_col = "Date",
                   parse_dates=True, na_values=['nan'])
```

```python
print(Data.head())
```

## Function

```python
def plot_data(symbol, title = "Historic Prices"):
    ax = symbol.plot(fontsize = 12)
    ax.set_xlabel("Data")
    ax.set_ylabel("Price")
    plt.show()
```

## Plot Data

```python
plot_data(Data)
```

```python
plot_data(Data['Close'])
```

## Daily Returns

```python
def compute_daily_returns(Symbol):
    dr = Symbol.copy()
    dr[1:] = (Symbol[1:]/Symbol[:-1].values)-1
    return dr
```

```python
Daily_Returns = compute_daily_returns(Data)
```

```python
print(Daily_Returns.head())
```

```python
plot_data(Daily_Returns["Close"][1:])
```

## Mean and Std. Dev

```python
mean = Daily_Returns['Close'].mean()
print(mean)
```

```python
std = Daily_Returns['Close'].std()
print(std)
```

## Histogram

```python
Daily_Returns['Close'][1:].hist(bins=100)
plt.axvline(mean, color = 'w', linestyle = 'dashed', linewidth = 2)
plt.axvline(std, color = 'r', linestyle = 'dashed', linewidth = 2)
plt.axvline(-std, color = 'r', linestyle = 'dashed', linewidth = 2)
```
